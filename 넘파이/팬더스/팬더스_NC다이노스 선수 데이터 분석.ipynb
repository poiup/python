# 1. 데이터 적재 및 통합

```python
# 라이브러리 가져오기
import pandas as pd
```

```python
# 데이터 가져오기
# read_excel() 기본동작 : 첫 번째 시트의 데이터만 가져오므로
data = pd.read_excel('data/NC Dinos.xlsx', sheet_name=None)
data
```

```python
# 시트별 데이터를 각각 개별 변수에 저장
nc13, nc14, nc15 = data.values()
```

```python
# nc13,14,15, 데이터컬럼의 Unnamed : 0 컬럼을 삭제해주세요
nc13.drop('Unnamed: 0', axis=1, inplace=True)
nc14.drop('Unnamed: 0', axis=1, inplace=True)
nc15.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
nc14.head(1)
```

```python
nc14.tail(1)
```

### 최종 데이터
2013,2014,2025 모두 합치되 각각 몇년도 데이터인지 알수있도록 연도값 추가

```python
#nc13, nc14,nc15 변수에 연도 컬럼을 추가
nc13['연도'] = 2013
nc14['연도'] = 2014
nc15['연도'] = 2015
```

```python
# 여러개의 데이터프레임 결합
# 결합방식  ; 컬럼개수와 속성이 완벽하게 동일하기때문에row개수만 늘어나도록 합니다.
# 최종결과 : 20row +20row +20row. 20columns
nc_result1 =  nc13.append([nc14,nc15], ignore_index=True)
```

```python
# append를 이용해 nc_result1 변수에 13,14,15년도 데이터를 합산해주세요
# nc13 : 0~19/ nc 14 : 20 ~ 39, nc15 40 ~59
nc_result2 = pd.concat([nc13, nc14, nc15], ignore_index=True)
```

```python
#nc_result2[0:40]
nc_result2[nc_result2['연도'] < 2015]
```

```python
#2015년 타율이 높은 상위 10명

#참고 ) 동일 타율을 기록한 선수에 대해서는 안타 개수가 많은 순서로 정렬
tmp = nc_result1[nc_result1['연도'] == 2015].sort_values(by=['타율','안타'], ascending=False).head(10)
tmp
```

```python
# 선수들만 보고싶다면 선수컬럼지정 + .tolist()를 이용합니다
tmp['선수명'].tolist()
```

```python
# pivot_table사용시
# index : 선수명
# values : 안타
# aggfunc : 평균
#tmp2 = nc_result1[nc_result1['연도'] <= 2015].sort_values(by=['타율','안타'], ascending=False)
#tmp2 = nc_result1.pivot_table(index='선수명', values='안타', aggfunc='mean').sort_values('안타', ascending=False)
tmp2 = nc_result1.pivot_table(index='선수명', values='안타', aggfunc='mean')
tmp2
```

## 4. 평균안타수 기준 상위 10위 타자들의 3년간 안타, 홈런,타율 조회

#### 참고) NaN값이 있을경우 '-'로 대체해서 표현

```python
nc_result1.set_index(['연도','선수명'])[['안타','홈런','타율']]
```

```python
# 1. 전체데이터 중에서 보고싶은 값을 기준으로 구조 변경 = > 해당 선수 데이터만 추출
tmp = nc_result1.set_index(['연도',
                           '선수명'])[['안타','홈런','타율']].unstack(0)

tmp = tmp.fillna('-')
tmp
```

```python
## 2. 전체 데이터 -> 보고싶은값 -> 구조변경 -> 해당 선수들의 데이터만 추출
# 보고싶은값 : 평균 안타수 기준 상위 10위 타자들의 3년간 '안타','홈런','타율'
# 추출방법 : 인덱싱/메서드
# tmp = nc_result1.set_index(['연도','선수명'])[['안타','홈런','타율']].unstack(0)

tmp2 = nc_result1.set_index(['연도'])[['선수명','안타','타율','홈런']].sort_values(by=['안타','타율','홈런'], ascending=False)
tmp
```

```python
# 안타갯수 상위 10위 선수들 목록을 뽑아주세요
# 피벗 테이블을 이용해 3년간 안타 총합이 많은순으로 나열해주시면 됩니다.
hit10 = nc_result1.pivot_table(index='선수명', values='안타', aggfunc='sum').sort_values(by='안타', ascending=False)
hit10
```

```python
hit10.head(10).index.tolist()
```

```python
#  fancy 인덱싱(조건색인에서 미리 준비된 리스트에 해당하는 목록만 인덱싱)
# 특정 데이터프레임에서 내가 제시한 명단에 해당하는 로우만 남기기
# .isin(리스트)
tmp3 = tmp2[tmp2['선수명'].isin(hit10.index.tolist())].head(10)
tmp3
```

```python
# 로우에 있던 연도를 컬럼으로 교환
tmp3 = tmp3.reset_index()
tmp3
```

```python
# 연도, 선수명으로 로우 설정
tmp3 = tmp3.set_index(['연도','선수명'])
```

```python
# unstack()을 이용해 연도를 컬럼으로 이동
# .fillna('-')로 NaN 처리
tmp3.unstack(0).fillna('-')
```
