### 2. 인기있는 영화찾기

- 기준 : 평점이 4.5 이상이고 평점 개수가 500개 이상인 영화

```python

data2.sort_values(by='평점', ascending=False).head(10)
```

```python
# 하나의 영화에 여러개의 평점이 존재하는 1:N관계이므로
# 사용자 아이디는 사실상 영화별 평균 평점을 구하는데 불필요한자료임
# 따라서 내가 원하는 로우(index) : 영화제목
# 원하는 컬럼(values) : 평점
# 집계함수(aggfunc) : 평균
# 으로 pivot_table을 data2에 걸어 집계해보세요. 정렬은 하지않아도 좋습니다
```

```python
data3 = data2.pivot_table(index='제목', values='평점', aggfunc='mean')
data3
```

```python
data3.sort_values(by='평점', ascending=False)
```

```python
# 평점을 기준으로 정렬하되 평균평점만 출력하는게 아니라 평점갯수도 같이 출력
# data2에 대해서 pivot_table을 걸때
# 로우 : 제목
# 값 : 평점
# 집계함수 : [mean,count]
result = data2.pivot_table(index='제목', values='평점', aggfunc=['mean','count'])
```

```python
result
```

```python
result.head(5)
```

```python
# result에 대해 swaplevel()을 이용해서 계층색인을 재배치해주세요
result2 =result.swaplevel(1,0, axis=1)
```

```python
# 평점  -> mean으로 정렬하되, 동점일경우 count가 클수록 위로
result2.sort_values(by=[('평점','mean'),('평점','count')] , ascending=False )
```

```python
# result3에 대해서 조건색인을 이용해
# 평점평균이 4.5 이상이면서 평점 개수가 500개 이상인 영화만 남겨보기
# ex) 이중으로 부여된 컬럼명은 result3['평점']['mean']과 같이 조회한다
# 변수 final_result에 저장해주세요
final_result = result2[(result2[('평점','count')] > 500) & (result2[('평점','mean')] >= 4.5)]
final_result.sort_values(by=[('평점','mean'),('평점','count')] , ascending=False )
```
