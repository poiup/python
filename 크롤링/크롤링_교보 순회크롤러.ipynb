```python
# 크롤링 필요 라이브러리 임포트
# 크롤링 작업을 위한 라이브러리 임포트

from bs4 import BeautifulSoup # 웹에서 가져온 데이터를 정제하는 역할

from selenium import webdriver #크롬창을 켜주는 역할

import requests # 파이썬에서 다이렉트로 자료를 요청하는 역할

import time # 코드중간에 딜레이를 넣는 역할
import codecs # 코덱스
```

```python
# browser라는 변수가 브라우저 그 자체, browser에 내리는 명령을 실제로 물리적인 드라이버가 수행함
browser = webdriver.Chrome("chromedriver.exe")
for i in range(1,21) :
    browser.get('https://www.aladin.co.kr/shop/common/wbest.aspx?BestType=Bestseller&BranchType=1&CID=0&page=%s&cnt=1000&SortOrder=1' % i)
    time.sleep(2)
    source += browser.page_source
    parsed_source = BeautifulSoup(source, "html.parser")
    ss_book_boxSource = parsed_source.find_all("div", class_="ss_book_box")
print(len(ss_book_boxSource))
   
```

```python
bookTitles = []
bookAuthors = []
bookPrices = []
for i in ss_book_boxSource :
    ss_book_infoSource = i.find_all("div", class_="ss_book_list")
    lis = ss_book_infoSource[0].find_all("li")
    if len(lis) > 4 :
        title = lis[1].text.split("|")[0].replace(","," ")
        author = lis[2].text.split("|")[0].replace(",","&")
        price = lis[3].text.split(" ")[0].replace(",","").replace("원","")
        title = title.split("ㅣ")[0].rstrip()
    else :
        title = lis[0].text.split("|")[0].replace(","," ")
        author = lis[1].text.split("|")[0].replace(",","&")
        price = lis[2].text.split(" ")[0].replace(",","").replace("원","")
        title = title.split("ㅣ")[0].rstrip()
    bookTitles.append(title)
    bookAuthors.append(author)
    bookPrices.append(price)
```

```python
for i in range(0,len(bookTitles)) : 
    print(bookTitles[i])
    print(bookAuthors[i])
    print(bookPrices[i])
    print("_----------------------_")
```

```python
f = codecs.open("C:/crawler1/aladin/bestS.txt", encoding="utf-8", mode="w")
f.write("제목, 글쓴이, 가격\n")
str = ""
for i in range(0,len(bookTitles)) : 
    str += (bookTitles[i] + ",")
    str += (bookAuthors[i]+ ",")
    str += (bookPrices[i]+ "\n")
f.write(str)
f.close()
```
